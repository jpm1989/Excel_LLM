````python
from transformers import AutoTokenizer, AutoModelForCausalLM,pipeline,set_seed
import torch
````

````python
model_name = "Qwen/Qwen3-4B"  # Replace with your local path if needed

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype=torch.bfloat16, trust_remote_code=True)
````

````python
pipe = pipeline('text-generation',model=model,tokenizer=tokenizer,
                     return_full_text=False,max_new_tokens=750,temperature=0.6, top_p=0.95, top_k=20, min_p=0)
````

````python
import os
import io
import matplotlib.pyplot as plt
import seaborn as sns
from contextlib import redirect_stdout
import sys
import re
````

````python

print(f"Loading Hugging Face model: {model_name}...")

# Determine device
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Load tokenizer and model

def ask_data_question_hf(dataframe: pd.DataFrame, question: str, hf_tokenizer, hf_model, hf_device: str):
    """
    Simulates Pandas AI's behavior using a locally loaded Hugging Face model.

    Args:
        dataframe (pd.DataFrame): The DataFrame to query.
        question (str): The natural language question.
        hf_tokenizer: Hugging Face tokenizer.
        hf_model: Hugging Face model.
        hf_device (str): Device to run inference on ('cuda' or 'cpu').

    Returns:
        tuple: (success (bool), result (str or error message), generated_code (str))
    """

    df_head = dataframe.head().to_string() # Provide schema context
    df_columns = dataframe.columns.tolist()

    # --- Prompt Engineering for Hugging Face Models ---
    # Many instruction-tuned models follow a chat-like format.
    # Adjust this based on the specific model's recommended prompting style.
    # For example, Mistral uses <s>[INST] instruction [/INST] model_response</s>
    # Phi-2 often works with simple prompts.

    # A general prompt for code generation
    prompt = f"""You are an expert Python programmer. Your task is to write a Python code snippet using the pandas library to answer a question about a DataFrame named `df`.

The DataFrame `df` has the following columns: {df_columns}
Here is a sample of the DataFrame:
{df_head}

Your response must be ONLY the Python code. Do not include any explanation, comments, or extra text.
The code should be executable and directly provide the answer to the question.
If the question asks for a plot, use `matplotlib.pyplot` and `seaborn` and save the plot to a file named 'generated_plot.png'. Ensure `plt.show()` is NOT called within the generated code.
If the question asks for a specific value or a small table, the last line of your code should be a `print()` statement or the final expression that evaluates to the answer.
If the question asks for a DataFrame or a larger output, the last line should be the DataFrame or the result of a DataFrame operation.
Only provide the output.
You are not required to read DataFrame(df) as it is already supplied to you.
- Incorrect: df = pd.read_csv("data.csv")

No need to "import" any python libraries.

Question: {question}

Python code:
"""

    # For instruction-tuned models, it's often best to wrap the prompt
    # according to their specific chat template.
    # This example uses a common template, adjust if your model has a different one.
    #if "instruct" in model_name.lower():
    messages = [
        {"role": "system", "content": "You are a helpful Python coding assistant that generates executable code."},
        {"role": "user", "content": prompt}
    ]
    # Some models (like Llama 2/Mistral Instruct) have specific chat formats
    # If the tokenizer has a chat template, use it.
    if hasattr(hf_tokenizer, 'apply_chat_template') and hf_tokenizer.chat_template:
          input_text = hf_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True,enable_thinking=False)
    else:
          # Fallback for models without a specific chat template or if you prefer a simpler instruction format
          input_text = f"<s>[INST] {prompt} [/INST]"
    #else:
        # For base models or simpler instruct models, a direct prompt might be fine
        #input_text = prompt

    #print(f"--- Sending prompt to LLM (Hugging Face) ---\n{input_text}\n")

    try:
        inputs = hf_tokenizer(input_text, return_tensors="pt", add_special_tokens=True).to(hf_device)

        # Generate code
        # Experiment with generation parameters for best results
        # num_beams > 1 for Beam Search (less creative, more focused)
        # do_sample=True for more diverse outputs
        # temperature for creativity (lower = less creative, higher = more creative)
        # max_new_tokens is crucial to limit output length
        # pad_token_id/eos_token_id are important for stopping generation correctly

        # Define generation_config or pass directly
        outputs = hf_model.generate(
            **inputs,
            max_new_tokens=250, # Max tokens for the generated code
            temperature=0.1,    # Low temperature for deterministic code
            do_sample=False,    # Do not sample for code generation
            num_return_sequences=1,
            # Common stop tokens for code generation
            eos_token_id=hf_tokenizer.eos_token_id,
            pad_token_id=hf_tokenizer.pad_token_id if hf_tokenizer.pad_token_id is not None else hf_tokenizer.eos_token_id,
            # Add custom stop strings (e.g., for markdown code blocks)
            # This is more complex with Hugging Face generate, often involves custom stopping criteria.
            # For simplicity, we'll rely on post-processing for now.
        )

        # Decode the generated text
        generated_full_text = hf_tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Extract only the generated code part
        # This requires robust parsing as LLMs can sometimes include conversational text
        # Look for the last occurrence of the prompt and extract what follows.

        # For instruction-tuned models, the response usually follows the prompt directly.
        # Find the start of the LLM's response part of the full generation.
        response_start_index = generated_full_text.find(input_text)
        if response_start_index != -1:
            generated_code = generated_full_text[response_start_index + len(input_text):].strip()
        else:
            generated_code = generated_full_text # Fallback if prompt not found in output

        # Clean up common markdown code block markers
        generated_code = re.sub(r'```python\n?', '', generated_code)
        generated_code = re.sub(r'\n?```.*', '', generated_code).strip() # Capture anything after the closing ```
        generated_code = generated_code.split('</think>')[-1]

        print(f"--- Generated Python Code (Hugging Face) ---\n{generated_code}\n")

        # --- 3. Secure Code Execution ---
        local_scope = {"df": dataframe, "pd": pd, "plt": plt, "sns": sns, "os": os, "io": io}

        output_capture = io.StringIO()
        with redirect_stdout(output_capture):
            exec(generated_code, {"__builtins__": {}}, local_scope)

        captured_output = output_capture.getvalue().strip()

        if "plt.savefig" in generated_code:
            plot_path = "generated_plot.png"
            if os.path.exists(plot_path):
                return True, f"Plot generated and saved to '{plot_path}'", generated_code
            else:
                return False, f"Code executed, but plot not found at '{plot_path}'. Output: {captured_output}", generated_code
        else:
            if captured_output:
                return True, captured_output, generated_code
            else:
                return True, "Code executed successfully (no explicit print output or plot generated).", generated_code

    except Exception as e:
        return False, f"Error executing code or generating response: {e}", generated_code
````

````python
q = "Which country has maximum genres and what are they?"
success, result, generated_code = ask_data_question_hf(df, q, tokenizer, model, device)

if success:
    print(f"Result:\n{result}")
    print(f"Generated code:\n{generated_code}")
else:
    print(f"FAILED: {result}")
print("\n" + "#"*70 + "\n")
````
